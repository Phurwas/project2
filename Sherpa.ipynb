## the most costly college and the cheapest college

```python
#combining annual cost and program to have one column of all institutionsd
df['COST']=df['COSTT4_A']
df['COST']=df['COST'].fillna(df['COSTT4_P'])
```

```python
print('The most costly college is:')
print(df.loc[df['COST'].idxmax()].INSTNM)
print('for the cost:')
print(df.loc[df['COST'].idxmax()].COST)
```

```python
print('The cheapest college is:')
print(df.loc[df['COST'].idxmin()].INSTNM)
print('for the cost:')
print(df.loc[df['COST'].idxmin()].COST)
```

## the average cost of colleges in different parts of US

```python
print('the average cost for each state')
print('by the ascending order')
print('')
print(classData.sort_values(ascending=True))
```

## the average cost of collge for religious vs secular institutions

```python
#creating a new dataframe of religius 
# droping all rows with NaN values
religiusCost= df.dropna(subset=['RELAFFIL'])
# calculation of average cost for religius schools
print('the average cost for religius institutions is:')
print(round(religiusCost.COST.mean()),'$')
```

```python
#creating a new dataframe of secular 
# droping all rows with NaN values
secularCost= df.loc[pd.isnull(df).any(1),:]
# calculation of average cost for secular schools
print('the average cost for secular institutions is:')
print(round(secularCost.COST.mean()),'$')
```

## Finding out the colleges which have an open admission policy

```python
# by using the data dictionary
# open admission policy indicatior is #1 yes #2 No
# creating dataframe for open admission institutions:
#droping rows with NaN values

Oadmission=df.dropna(subset=['OPENADMP'])
# drop rows with non open admission
Oadmission= Oadmission[Oadmission.OPENADMP !=2]

#calculationg the percentage of open admission of all colleges
print('There are', len(Oadmission),'Collegs with open admission')
print('which is', round((len(Oadmission))/(len(df))*100),'% of',len(df),'colleges')
```

## colleges which have highest and lowest family income and how does that correlate with college costs

```python
df['FAMINC']=df['FAMINC'].replace('PrivacySuppressed',np.nan)
df['FAMINC']=df.FAMINC.astype(float)
```

```python
print('the maximum family income is: ',round(df['FAMINC'].max()),'$')
print('College:',df.iloc[df['FAMINC'].idxmax].INSTNM,',',df.iloc[df["FAMINC"].idxmax].STABBR)
```

```python
print('the minimum family income is: ',round(df['FAMINC'].min()),'$')
print('College:',df.iloc[df['FAMINC'].idxmin].INSTNM,',',df.iloc[df["FAMINC"].idxmin].STABBR)
```
